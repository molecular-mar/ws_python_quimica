Equilibrio químico: encontrar el pH de una solución de bicarbonato.

```python
import chempy
from collections import defaultdict
from chempy.equilibria import EqSystem
from math import log10

#Definimos un sistema de ecuaciones químicas
eqsys = EqSystem.from_string("""HCO3- = H+ + CO3-2; 10**-10.3
  H2CO3 = H+ + HCO3-; 10**-6.3
  H2O = H+ + OH-; 10**-14/55.4
""")  # pKa1(H2CO3) = 6.3 , pKa2=10.3 & pKw=14

#Resolvemos el sistema de ecuaciones
arr, info, sane = eqsys.root(defaultdict(float, {'H2O': 55.4, 'HCO3-': 1e-2}))

conc = dict(zip(eqsys.substances, arr))


print("pH: %.2f" % -log10(conc['H+']))
```

Cinética Química

```python
from chempy import ReactionSystem  # Las velocidades utilizadas son arbitrarias
from chempy.kinetics.ode import get_odesys #ode -> ordinary differential equations
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt


#Definimos el sistema de reacciones
rsys = ReactionSystem.from_string("""2 Fe+2 + H2O2 -> 2 Fe+3 + 2 OH-; 42
    2 Fe+3 + H2O2 -> 2 Fe+2 + O2 + 2 H+; 17
    H+ + OH- -> H2O; 1e10
    H2O -> H+ + OH-; 1e-4""")  # "[H2O]" = 1.0 (en realidad, 55.4 a temperatura ambiente)

#Obtemos el sistema de ecuaciones diferenciales
odesys, extra = get_odesys(rsys)

tout = sorted(np.concatenate((np.linspace(0, 23), np.logspace(-8, 1)))) # Intervalo grande/chico
c0 = defaultdict(float, {'Fe+2': 0.05, 'H2O2': 0.1, 'H2O': 1.0, 'H+': 1e-2, 'OH-': 1e-12})
result = odesys.integrate(tout, c0, atol=1e-12, rtol=1e-14) #integra el sistema de EDOs

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax in axes:
    _ = result.plot(names=[k for k in rsys.substances if k != 'H2O'], ax=ax)
    _ = ax.legend(loc='best', prop={'size': 9})
    _ = ax.set_xlabel('Tiempo')
    _ = ax.set_ylabel('Concentración')
_ = axes[1].set_ylim([1e-13, 1e-1])
_ = axes[1].set_xscale('log')
_ = axes[1].set_yscale('log')
_ = fig.tight_layout()
_ = fig.savefig('kinetics.png', dpi=72)
```

## ASE: Atomic Simulation Environment

Para crearr, manipular, ejecutar, visualizar y analizar simulaciones atomisticas.

```python
!pip install ase
```

```python
#Energia de atomización de N2 usando EMT
from ase import Atoms
from ase.calculators.emt import EMT

atom = Atoms('N')
atom.calc = EMT()
e_atom = atom.get_potential_energy()

d = 1.1
molecule = Atoms('2N', [(0., 0., 0.), (0., 0., d)])
molecule.calc = EMT()
e_molecule = molecule.get_potential_energy()

e_atomization = e_molecule - 2 * e_atom

print('Nitrogen atom energy: %5.2f eV' % e_atom)
print('Nitrogen molecule energy: %5.2f eV' % e_molecule)
print('Atomization energy: %5.2f eV' % -e_atomization)
```

Dinámica molecular

Creamos un cristal de Cu, con momentos correspondiendo a 300 K, y luego integramos la segúnda ley de Newton de forma numérica con un tamaño de paso de 5 femtosegundos. 

```python
"""Dinamica molecular a energía constante (NVE, microcanónico)."""

from ase.lattice.cubic import FaceCenteredCubic
from ase.md.velocitydistribution import MaxwellBoltzmannDistribution
from ase.md.verlet import VelocityVerlet
from ase.io.trajectory import Trajectory
from ase import units

# Use Asap for a huge performance increase if it is installed
use_asap = False

if use_asap:
    from asap3 import EMT
    size = 10
else:
    from ase.calculators.emt import EMT
    size = 3

# Establecemos el cristal
atoms = FaceCenteredCubic(directions=[[1, 0, 0], [0, 1, 0], [0, 0, 1]],
                          symbol='Cu',
                          size=(size, size, size),
                          pbc=True)

# Para las interacciones interatómicas usamos Effective Medium Theory
atoms.calc = EMT()

# Establecemos los momentos correspondientes a T=300K
MaxwellBoltzmannDistribution(atoms, temperature_K=300)

# Usaremos el algoritmo de dinámica molecular VelocityVerlet.
dyn = VelocityVerlet(atoms, 5 * units.fs)  # Pasos de 5 fs.


def printenergy(a=atoms):
    """Función para imprimir energía potencial, cinética y total"""
    epot = a.get_potential_energy() / len(a)
    ekin = a.get_kinetic_energy() / len(a)
    print('Energía por átomo: Epot = %.3feV  Ecin = %.3feV (T=%3.0fK)  '
          'Etot = %.3feV' % (epot, ekin, ekin / (1.5 * units.kB), epot + ekin))


# Podemos integrar al cálculo la función de imprimir el resultado cada 20
# pasos
dyn.attach(printenergy, interval=20)

# También vamos registrando las trayectorias.
traj = Trajectory('moldyn3.traj', 'w', atoms)
dyn.attach(traj.write, interval=50)

# Now run the dynamics
printenergy()
dyn.run(500)
```

Deep learning puede ser utilizado para resolver muchos tipos de problemas, pero el flujo de trabajo básico es usuelmente el mismo:

1. Seleccionar o crear los datos con los que entrenarás un modelo.
1. Crear un modelo.
1. Entrena el modelo con un conjunto de entrenamiento.
1. Evalua el modelo con un conjunto de prueba para ver que tal funciona.
1. Usa el modelo para realizar predicciones.



Vamos a intentar predecir la solubilidad de moleculas pequeñas dadas sus fórmulas químicas. Ésta es una propiedad muy importante en el desarrollo de medicamentos: si un medicamento no es suficientemente soluble, probablemente no podra disolverse suficiente en la corriente sanguinea para poder tener un efecto terapéutico.

Lo primero que necesitamos en un conjunto de datos de solubilidades medidas para moleculas reales.Uno de los componente de DeepChem es MoleculeNet, una collección diversa de datos químicos y moleculares. Usaremos quí el conjunto de solubilidad de Delaney. La solubilidad en este conjunto de datos está reportada como log(solubilidad) con la solubilidad en mol/L.

```python
tasks, datasets, transformers = dc.molnet.load_delaney(featurizer='GraphConv')
train_dataset, valid_dataset, test_dataset = datasets
```



No entraremos en mucho detalle, pero nota el argumento de featurizer. Las moleculas pueden representarse de muchas formas, y aquí decimos que representación queremos usar, o cómo queremos *describir* (featurize) los datos. Nota que obtenemos tres conjuntos de datos: un conjunto de entrenamiento, un conjunto de validación, y un conjunto de prueba. Cada uno tiene una función en el flujo de trabajo estándar de Deep Learning.

Ahora que tenemos los datos, el siguiente paso es crear un modelo. Usaremos un tipo particular llamado "red convolucional gráfica" o "graphconv".


```python
model = dc.models.GraphConvModel(n_tasks=1, mode='regression', dropout=0.2)
```

Ahora necesitamos entrenar nuestro modelo. Para ello necesitamos indicarle el conjunto de datos de entrenamiento y decirle cuantos periodos (epochs) de entrenamiento debe realizar (es decir, cuantas revisiones completas de nuestros datos).

```python
model.fit(train_dataset, nb_epoch=100)
```

Si todo salió bien, ahora tenemos un modelo entrenado. Ahora debemos evaluar si funciona correctamente seleccionando una métrica de evaluación y evaluando el modelo. En este ejemplo usaremos como métrica el coeficiente de correlación de Pearson r2.Podemos evaluarlo tanto en el conjunto de entrenamiento como en el de prueba.

```python
metric = dc.metrics.Metric(dc.metrics.pearson_r2_score)
print("Puntaje conjunto de entrenamiento:", model.evaluate(train_dataset, [metric], transformers))
print("Puntaje conjunto de prueba:", model.evaluate(test_dataset, [metric], transformers))
```



Nota que obtenemos un buen puntaje para el conjunto de prueba, lo cuál suele ocurrir como podrías esperar. Cuando el puntaje para es muy bueno para el conjunto de entrenamiento pero malo para datos externos se dice que realizamos un *sobreajuste* o "overfitting"; es muy importante siempre evaluar el desempeño del modelo.

En nuestro caso, nuestro modelo ajustado es relativamente bueno, y lo usaremos para hacer predicciones. Probemos con las primeras 10 moléculas del conjunto de prueba. Para cada una imprimamos su estructura química (representada por una cadena SMILES) y el valor de log(solubilidad) predicho y real.


```python
# Para visualizar las moleculas usaremos RDKit
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
solubilities = model.predict_on_batch(test_dataset.X[:10])
moleculas_lista = []
for molecule, solubility, test_solubility in zip(test_dataset.ids, solubilities, test_dataset.y):
    print(solubility, test_solubility, molecule)
    moleculas_lista.append(Chem.MolFromSmiles(molecule))
```

```python
# Para ver las moléculas
img_moleculas = Draw.MolsToGridImage(moleculas_lista, molsPerRow=2)
img_moleculas
```
